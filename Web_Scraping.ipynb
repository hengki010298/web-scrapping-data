```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL=["https://manadoline.com/?s=cap+tikus","https://manadoline.com/page/2/?s=cap+tikus"]


titles = []
contents = []
for url in URL:
    link = requests.get(url)
    soup = BeautifulSoup(link.text, "html.parser")
    for title , content in zip(soup.select('.entry-title'),soup.select('.td-post-date')):
        titles.append(title.text.strip())
        contents.append(content.text.strip())
df5 = pd.DataFrame({"Title":titles,"Date":contents})
print(df5)
```

```python
df6 = pd.concat([df,df2,df3,df4,df5])
print (df6)
```

```python
# here is texblob our natural lanaguage processing library
from textblob import TextBlob
# let's create a function that bring back sentiment and translations
def translation(text):
  blob =TextBlob(text)
  return str(blob.translate(from_lang='id', to='en'))

def sentiment(text):
  blob=TextBlob(text)
  return blob.sentiment.polarity

df6['translation'] = df6['Title'].apply(translation)
df6['sentiment'] =df6['translation'].apply(sentiment)
```
